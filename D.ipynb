```python
p = figure(plot_width=400,plot_height=300,title='Bokeh Example',
          x_axis_label='X',y_axis_label='Y')
x = np.linspace(-2,2,30)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
p.line(x,y1,legend='sin(x)',color = 'blue')
p.circle(x,y2,legend='cos(x)',fill_color='green',size=5)
p.triangle(x,y3,legend='sin(x)+cos(x)',color = 'red')
show(p)
```

```python
from bokeh.layouts import gridplot
```

```python
x = np.linspace(-2,2,30)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
s1 = figure(plot_width=300,plot_height=200)
s1.circle(x,y1,size=10,color='navy',alpha = 0.7)

s2 = figure(plot_width=300,plot_height=200)
s2.triangle(x,y2,size=10,color='firebrick',alpha = 0.7)

s3 = figure(plot_width=300,plot_height=200)
s3.square(x,y3,size=10,color='olive',alpha = 0.7)
p = gridplot([[s1,s2,s3]])
show(p)
```

```python
import plotly.graph_objs as go
```

```python
Z = np.linspace(0,10,50)
X = np.cos(Z)
Y = np.sin(Z)
trace = go.Scatter3d(x=X,y=Y,z=Z,
                    mode = 'markers',
                    marker=dict(size=12,color=Z,colorscale='Viridis'))
layout=go.Layout(title="3D interactive plot")
fig = go.Figure(data=[trace],layout=layout)
fig.show()
```

```python
X,Y = np.meshgrid(np.linspace(-2,2,30),np.linspace(-2,2,30))
Z = np.cos(X**2 + Y**2)
trace = go.Surface(x=X,y=Y,z=Z)
go.Figure(data = [trace],layout=go.Layout(title='3D Surface')).show()
```
