```python
fig, ax = plt.subplots(1,3)
img2 = img.copy()
img2[:,:,2] = img[:,:,2]*5
ch = ['Red','Green','Blue']
for i,axi in enumerate(ax.flat):
    axi.imshow(img2[:,:,i],plt.cm.gray)
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch[i],color=ch[i].lower())
    fig.set_size_inches(20,50)
```
