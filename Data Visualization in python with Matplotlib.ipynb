# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [1,2,3,4,]
y = [2,4,6,8,]

# Resize your Graph
plt.figure(figsize=(5,3), dpi=100)

# line 1

#plt.plot(x,y, label= '2x', color='blue', linewidth=2, marker= '.',linestyle='--', markersize=10, markeredgecolor='red')

# use shorthand notation
# fnt = '[color][marker][line]'

plt.plot(x,y, 'b^--', label='2x')

# line number two

# select interval we want to plot point at
x2 = np.arange(0,4.5,0.5)
print(x2)

# plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label= 'x^2')

# plot remainder of graph as a dpi
plt.plot(x2[5:], x2[5:]**2, 'r--')

# add a title (specify font parameters with fontdict)
plt.title('0ur First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# x and y labels
plt.xlabel('X axis', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y axis')

# x, y axis Tichmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

#plt.bar(labels, values)

plt.figure(figsize=(6,4))

plt.show()
```
